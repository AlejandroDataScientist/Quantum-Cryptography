```python
import numpy as np
from scipy import stats
import random
```

```python
def frecuencia_calculo(Fini,Ch,TT):
  F = np.zeros(20)
  F[0] = 4e6
  G = (Ch==1)
  T1=TT[G]
  N1=len(T1)
  pt = 81e-12
  DT1=np.zeros(N1-1)
  ts = np.zeros(20)
  for s in range(19):
    St=1/(F[s]*pt)
    u=DT1
    for i in range(N1-1):
        DT=(T1[i+1]-T1[i])
        u[i]=np.round(DT/St)
        if u[i]== 1:
            DT1[i]=DT
        else:
            DT1[i]=DT-(u[i]-1)*St;

    DT1=np.sort(DT1)
    N=len(DT1)
    i=0
    k=0
    t1 = np.empty(N)
    t1[:] = np.nan
    C1 = t1
    while i < N:
        t1[int(k)]=DT1[int(i)]
        C1[int(k)]=sum(t1[int(k)]==DT1)
        i=i+C1[k]
        k=k+1
    t1 = t1[~np.isnan(t1)]
    C1 = C1[~np.isnan(C1)]
    t1r=t1*pt
    dC1=C1/np.sum(C1)
    tm=np.sum(t1r*dC1)
    ts[s]=np.sqrt(sum(((t1r-tm)*(t1r-tm))*dC1))
    F2=1/tm
    F[s+1]=(F[s]+F2)/2
    if s>1:
        if ts[s] >= ts[s-1]:
            break
  F = np.trim_zeros(F)
  print(F[-1])
  return
```
